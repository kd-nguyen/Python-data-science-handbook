## Text and annotation

```python
births = pd.read_csv('births.csv')
```

```python
quartiles = np.percentile(births['births'], [25, 50, 75])
quartiles
```

```python
mu = quartiles[1]
sig = 0.75 * (quartiles[2] - quartiles[0])
```

```python
births_capped = births.query('(births > @mu - 5 * @sig) & (births < @mu + 5 * @sig)')
births_capped.head()
```

```python
births_capped.loc[:, 'day'] = births_capped['day'].astype(int)
```
